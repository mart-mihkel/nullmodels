```python
import numpy as np
import pandas as pd

from plotnine import *
from nullmodels import *
```

```python
n = 25
p_ub, r_ub, th = pr_quantile_hypergeom(n)
p_lb, r_lb, _  = pr_quantile_hypergeom(n, q=0.1)

df = pd.DataFrame({'precision': np.concatenate((p_ub, p_lb)),
                   'recall': np.concatenate((r_ub, r_lb)),
                   'threshold': np.tile(th, 2)})

p  = ggplot(df)
p += aes('recall', 'precision', color='threshold', group='threshold')
p += geom_line()
p
```
