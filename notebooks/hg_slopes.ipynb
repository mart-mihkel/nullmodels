```python
import numpy as np
import pandas as pd

from plotnine import *
from nullmodels import *
```

```python
n = 25
p_ub, r_ub, th = pr_quantile_hypergeom(n, q=.9)
p_lb, r_lb, _  = pr_quantile_hypergeom(n, q=.1)

df = pd.DataFrame({'Precision': np.concatenate((p_ub, p_lb)),
                   'Recall': np.concatenate((r_ub, r_lb)),
                   'Threshold': np.tile(th, 2)})

(ggplot(df) +
 aes('Recall', 'Precision', color='Threshold', group='Threshold') +
 geom_line() +
 ggtitle(r'Hypergeom $\bar{q}_{0.9}$ to $q_{0.9}$ per classification threshold') +
 theme_bw())
```
